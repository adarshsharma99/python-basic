5. What is mutability in Python
- Mutability in Python
If you're referring to mutability, it's a fundamental concept in Python:

Mutability refers to the ability of an object to change its state after creation.

Mutable Objects
In Python, some objects are mutable, meaning their contents can be modified after creation. Examples include:

1. Lists: [1, 2, 3]
2. Dictionaries: {"name": "John", "age": 30}
3. Sets: {1, 2, 3}

Immutable Objects
Other objects are immutable, meaning their contents cannot be modified after creation. Examples include:

1. Integers: 5
2. Strings: "hello"
3. Tuples: (1, 2, 3)

Understanding mutability is crucial in Python programming, as it affects how objects are passed around and modified in your code.

7. What is the difference between “==” and “is” operators in Python
- ## "==" Operator
The == operator in Python checks for equality between two values. It returns True if the values are equal and False otherwise.

Example:
a = 5
b = 5
print(a == b)  # Output: True
## "is" Operator
The is operator in Python checks for identity between two objects. It returns True if both variables refer to the same object in memory and False otherwise.

Example:
a = [1, 2, 3]
b = [1, 2, 3]
print(a == b)  # Output: True (values are equal)
print(a is b)  # Output: False (not the same object in memory)
## Key Differences
1. Equality vs Identity: == checks for equality of values, while is checks for identity of objects.
2. Memory Location: is checks if both variables point to the same memory location, whereas == checks if the values are equal, regardless of memory location.

12. How does the elif statement work
- Elif Statement in Python*

The elif statement is used to check another condition if the initial if condition is false. Here's a breakdown of how it works:

Syntax
if condition1:
    # code to execute if condition1 is true
elif condition2:
    # code to execute if condition1 is false and condition2 is true
How it Works
1. Initial condition evaluation: The if condition is evaluated.
2. Elif evaluation: If the if condition is false, the elif condition is evaluated.
3. Code execution: If the elif condition is true, the code indented under the elif statement is executed.

Example
x = 5
if x > 10:
    print("x is greater than 10")
elif x == 5:
    print("x is equal to 5")
In this example:

1. The if condition x > 10 is evaluated and found to be false.
2. The elif condition x == 5 is evaluated and found to be true.
3. The code print("x is equal to 5") is executed.

13. 8 What is the difference between for and while loops
- For and while
Key Differences
1. Iteration: for loops iterate over a sequence, while while loops iterate based on a condition.
2. Termination: for loops terminate when the sequence is exhausted, while while loops terminate when the condition becomes false.
3. Control: for loops provide less control over the iteration process, while while loops provide more control.

14. Describe a scenario where a while loop is more suitable than a for loop.
- ## Scenario: User Input Validation
Imagine you're writing a program that asks the user to input a password. You want to ensure that the password meets certain criteria, such as having a minimum length. You can use a while loop to repeatedly ask the user for input until they enter a valid password.

### Example
password = input("Enter a password (min 8 characters): ")
while len(password) < 8:
    print("Password is too short. Please try again.")
    password = input("Enter a password (min 8 characters): ")
print("Password accepted!")
## Why While Loop is More Suitable
1. Unknown number of iterations: You don't know how many times the user will need to enter a password before it meets the criteria.
2. Conditional termination: The loop terminates when the user enters a valid password, which is a condition that needs to be evaluated.

In this scenario, a while loop is more suitable than a for loop because it allows you to repeatedly execute a block of code based on a condition, rather than iterating over a fixed sequence.
